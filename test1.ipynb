```python
import matplotlib.pyplot as plt
import cv2
import mtcnn
```

```python
img = plt.imread('test1.jpg')
plt.imshow(img)
```

```python
# draw an image with detected objects
def draw_image_with_boxes(filename, result_list):
    data = plt.imread(filename)
    for result in result_list:
        x, y, w, h = result['box']
        x2, y2 = x + w, y + h
        cv2.rectangle(img, (x, y), (x2, y2), (0,0,255), 1)
        for key, value in result['keypoints'].items():
            cv2.circle(img,center=value, radius=2, color=(0,255,0),thickness=-1)
    plt.figure(figsize=(10,12))
    plt.imshow(img)
    plt.show()
```

```python
detector = mtcnn.MTCNN()
# detect faces in the image
faces = detector.detect_faces(img)
# display faces on the original image
draw_image_with_boxes('test2.jpg', faces)
```

```python
classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img = plt.imread('test1.jpg')
faces = classifier.detectMultiScale(img,1.4,10)
faces
```

```python
for face in faces:
    x,y,w,h = face
    x2, y2 = x + w, y + h
    #draw a rectangle over the pixels
    cv2.rectangle(img, (x, y), (x2, y2), (0,0,255), 1)
plt.figure(figsize=(10,12))
plt.imshow(img)
plt.show()
```
