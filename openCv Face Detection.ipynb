openCV provides a number of pre-trained models as part of the installation. These are available on your system and are also available on the https://github.com/opencv/opencv/tree/master/data/haarcascades

```python
classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
import matplotlib.pyplot as plt
```

```python
img = plt.imread('test2.jpg')
# img = cv2.imread('test2.jpg')
```

```python
# cv2.imshow('Image in Consideration',img)
# # keep the window open until we press a key
# cv2.waitKey(0)
# # close the window
# cv2.destroyAllWindows()
plt.imshow(img)
```

```python
#classifier.detectMultiScale(image,scaleFactore,minNeighbours)
"""
The scale factor controls how the input image is scaled prior to detector.
e.g. is it scaled up or down, which can help to better find the faces in the image.
(default 1.1)

minNeighbours determine how robust each detection must be in order to be reported
e.g. the number of candidate rectangles that found the face.
increase to minNeighbour value require a lot more confidence before a face is detected.
(default 3)
"""
faces = classifier.detectMultiScale(img,1.05,8)
```

```python
for face in faces:
    print(face)
```

```python
for face in faces:
    x,y,w,h = face
    x2, y2 = x + w, y + h
    #draw a rectangle over the pixels
    cv2.rectangle(img, (x, y), (x2, y2), (0,0,255), 1)
plt.figure(figsize=(10,12))
plt.imshow(img)
plt.show()
```
